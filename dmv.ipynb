```python
def get_sample(df,columns):

    result = {}
    for col in columns:
        result[col] = df[col][:5]

    return result


```

```python
import numpy as np
churn_df.replace(' ',np.nan,inplace=True)
churn_df.replace('?',np.nan,inplace=True)
```

```python
churn_df.isna().sum()
```

```python
churn_df['TotalCharges'] = pd.to_numeric(churn_df['TotalCharges'])
churn_df['TotalCharges'].fillna(churn_df['TotalCharges'].mean(),inplace=True)
```

```python
numeric_col = []

for col in columns:
    if churn_df[col].dtype in ['int64' ,'float64']:
        numeric_col.append(col)

numeric_col = numeric_col[1:]
numeric_col
```

```python
def show_outliers(df,columns):
    plt.figure(figsize=(10,6 * len(columns)))

    for i,col in enumerate(columns):
        plt.subplot(len(columns),1,i+1)
        plt.boxplot(df[col])
        plt.title(f'boxplot of {col}')
        plt.xlabel(col)

    plt.tight_layout()
    plt.show()
```

```python
show_outliers(churn_df,numeric_col)
```

```python
churn_df['Tenure_years'] = churn_df['tenure'] / 12
churn_df['avg_charges_per_month'] = churn_df['TotalCharges'] / churn_df['tenure']
```

```python
churn_df.head()
```

```python
from sklearn.model_selection import train_test_split

columns = churn_df.columns

numeric_columns = []

for col in columns:
    if churn_df[col].dtype in ['int64','float64']:
        numeric_columns.append(col)

numeric_columns
```

```python
churn_df.drop(columns='customerID',inplace=True)
columns = churn_df.columns
```

```python
cat_columns = [col for col in columns if col not in numeric_columns]
cat_columns
```

```python
from sklearn.preprocessing import LabelEncoder

X = churn_df.drop(columns='Churn')
y = churn_df['Churn']

label = LabelEncoder()

for col in cat_columns[:-1]:
    X[col] = label.fit_transform(X[col])
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

```python
for i,value in enumerate(real_estate_df['total_sqft']):
    tokens = str(value).split('-')

    if len(tokens) == 2:

        val = (float(tokens[0]) + float(tokens[1])) / 2
        real_estate_df['total_sqft'].replace(value,int(val),inplace=True)


    if 'Sq. Meter' in str(value):

        index = str(value).find('Sq. Meter')
        val = value[:index]
        val = float(val)*10.7
        real_estate_df['total_sqft'].replace(value,int(val),inplace=True)

    if 'Sq. Yards' in str(value):

        index = str(value).find('Sq. Yards')
        val = value[:index]
        val = float(val)*9
        real_estate_df['total_sqft'].replace(value,int(val),inplace=True)

#Sq. Meter
#Sq. Yards
```

```python
unwanted_units = ['Perch', 'Acres', 'Guntha', 'Cents', 'Grounds']


for i,value in real_estate_df['total_sqft'].items():

    if any(unit in str(value) for unit in unwanted_units):
        real_estate_df.drop(index=i,inplace=True)


real_estate_df.reset_index(drop=True,inplace=True)
```

```python
import requests
import pandas as pd
```

```python
lat = 19.0211
lon = 72.5134
```

```python
api_key = "14b01e724014c68826132595f1214a55"
api_url = f"http://api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid={api_key}"
```

```python
response = requests.get(api_url)
weather_data = response.json()
weather_data
```

```python
temperature =  [item['main']['temp'] for item in weather_data['list']]
temperature[-1]
```

```python
humidity =  [item['main']['humidity'] for item in weather_data['list']]
humidity[-1]
```

```python
wind_speed = [item['wind']['speed'] for item in weather_data['list']]
wind_speed[-1]
```

```python
weather_desc = [item['weather'][0]['description'] for item in weather_data['list']]
weather_desc[-1]
```

```python
date_time = [item['dt_txt'] for item in weather_data['list']]
date_time[-1]
```

```python
weather_df = pd.DataFrame({'date_time':date_time,
                          'temperature':temperature,
                          'humidity':humidity,
                          'wind_speed':wind_speed,
                          'weather_description':weather_desc})


weather_df.head()
```

```python
weather_df['date_time'] = pd.to_datetime(weather_df['date_time'])
weather_df['date'] = weather_df['date_time'].dt.date
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.lineplot(weather_df,x='date_time',y='temperature',label='temp')
sns.lineplot(weather_df,x='date_time',y='wind_speed',label='wind speed')
sns.lineplot(weather_df,x='date_time',y='humidity',label='humidity')
plt.legend()
plt.show()
```

```python
print(weather_df['temperature'].mean())
weather_df['temperature'].max(),weather_df['temperature'].min()
```

```python
sns.scatterplot(x=weather_df['temperature'],y=weather_df['humidity'])
```

```python
weather_df.set_index(weather_df['date_time'],inplace=True)
```

```python
daily_mean_temp = weather_df['temperature'].resample('D').mean()
daily_mean_humidity = weather_df['humidity'].resample('D').mean()
daily_mean_wind_speed = weather_df['wind_speed'].resample('D').mean()
```
